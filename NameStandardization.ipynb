```python
#RUN WHOLE SCRIPT STEP BY STEP

import pandas as pd

df = pd.read_csv("Main_clean.csv")
init_shape = df.shape
print("Initial Shape ", init_shape)
print("Column names:")
for col in df.columns:
    print(col)
```

```python
#COLUMN RENAME FOR BETTER READABILITY

pretty_cols = {}
for col in df.columns:
    new = col

    # remove prefix
    new = new.replace("Graduate_School_Dashboard_Dataset_", "")

    # fix typos
    new = new.replace("Citizensihp", "Citizenship")

    # convert to readable form
    new = new.replace("_", " ").strip()
    new = " ".join(word.capitalize() for word in new.split())

    pretty_cols[col] = new
print("Updated Column Names:")
for old, new in pretty_cols.items():
    print(f"{old} --> {new}")
df.rename(columns=pretty_cols, inplace=True)
    
```
